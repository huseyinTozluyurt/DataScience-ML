```python
myFrame=pd.DataFrame(data_frame, columns=['Name','Wage','Value'])
myFrame
```

```python
def value_to_float(x):
    if type(x)==float or type(x)==int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K',''))*1000
    if 'M' in x:
        if len(x) >1:
            return float(x.replace('M',''))*1000000
    if 'B' in x:
        if len(x) >1:
            return float(x.replace('B',''))*1000000000
wage = myFrame['Wage'].replace('[/€,]','',regex=True).apply(value_to_float)
value = myFrame['Value'].replace('[/€,]','',regex=True).apply(value_to_float)

myFrame['Wage']=wage
myFrame['Value']=value

myFrame['difference'] = myFrame['Value'] - myFrame['Wage']
myFrame
```

```python
myFrame.sort_values(by=['difference'], ascending=True, inplace=True)
myFrame
```

```python
graph=sns.scatterplot(x='Wage', y='Value', data=myFrame)
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
X=iris.data
y=iris.target
```

```python
X
```

```python
y
```

```python
feature_names=iris.feature_names
target_names=iris.target_names
```

```python
target_names
```

```python
print(type(X))
print(type(y))
print(type(feature_names))
print(type(target_names))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=42)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred))
```

```python
sample = [[3,5,4,2],[2,3,5,4]]
predictions = knn.predict(sample)
pred_species = [iris.target_names[p] for p in predictions]
print("Predictions: ",pred_species)
```
